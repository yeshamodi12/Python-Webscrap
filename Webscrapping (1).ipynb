```python
import numpy as np
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Lists_of_Hindi_films'
Web_page = requests.get(url)
soup = BeautifulSoup(Web_page.text,"html.parser")

```

```python
soup
```

```python
data = soup.find_all('table')[1]
```

```python
print(data)
```

```python
movie_titles = data.find_all('th')
```

```python
movie_titles
```

```python
movie_titles_data = [title.text.strip() for title in movie_titles]

print(movie_titles_data)
```

```python
data_columm = pd.DataFrame(columns=movie_titles_data)
```

```python
data_columm
```

```python
raw_data = data.find_all('tr')
raw_data
```

```python
for row in raw_data[1:]:
    row_elements = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_elements]

    # Check if the number of elements in individual_row_data matches the number of columns in df
    if len(individual_row_data) == len(data_columm.columns):
        length = len(data_columm)
        data_columm.loc[length] = individual_row_data
    else:
        print("Number of elements in individual_row_data doesn't match the number of columns in df")
```
